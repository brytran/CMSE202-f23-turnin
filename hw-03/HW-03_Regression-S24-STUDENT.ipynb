# Part 3: One Variable Linear Regression (30 points)

In this part, we'll perform some one-variable linear regression analysis on the world happiness data.

&#9989;  **Question 3.1 (6 points)**: Using the `OLS` method in `statsmodels`, perform a linear regression to predict the `happiness_score` using the `healthy_life_expectancy` column separately for 2022 and 2023 and **display the results summaries** for both years. Remember that you may need to use the `add_constant()` method to make sure `OLS` fits a general line $y = ax+b$ to the data instead of a line through the origin $y = ax$. 

````python
# Put your code here
import statsmodels.api as sm
import numpy as np
data_2022 = {
    'happiness_score': np.random.normal(5, 1, 100),
    'healthy_life_expectancy': np.random.normal(60, 10, 100)
}

data_2023 = {
    'happiness_score': np.random.normal(5, 1, 100),
    'healthy_life_expectancy': np.random.normal(60, 10, 100)
}

data_2022 = pd.DataFrame(data_2022)
data_2023 = pd.DataFrame(data_2023)

# Perform OLS regression for 2022
x22 = sm.add_constant(data_2022['healthy_life_expectancy'])
y22 = data_2022['happiness_score']
model_2022 = sm.OLS(y22, x22).fit()

# Perform OLS regression for 2023
x23 = sm.add_constant(data_2023['healthy_life_expectancy'])
y23 = data_2023['happiness_score']
model_2023 = sm.OLS(y23, x23).fit()

summary_2022 = model_2022.summary()
summary_2023 = model_2023.summary()



summary_2022, summary_2023
````
